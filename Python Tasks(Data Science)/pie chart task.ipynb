```python
import matplotlib.pyplot as plt
```

```python
#### **1. Create a Basic Pie Chart**  
#Write a Python script to create a **pie chart** with the following categories:  
#- Categories: `['Apple', 'Banana', 'Cherry', 'Grapes']`  
#- Values: `[30, 20, 40, 10]` 

plt.figure(figsize=(12, 8))  # Adjust figure size for better visualization
plt.subplot(3, 3, 1)
categories = ['Apple', 'Banana', 'Cherry', 'Grapes']
values = [30, 20, 40, 10]
plt.pie(values, labels=categories)
plt.title('Basic Pie Chart')
```

```python
#### **2. Add Percentage Labels to a Pie Chart**  
#Modify the **pie chart** to display **percentage values** inside each slice using `autopct='%1.1f%%'`. 

plt.subplot(3, 3, 2)
plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title('Pie Chart with Percentages')
```

```python
#### **3. Customize Slice Colors**  
#Assign specific colors to each slice of the **pie chart** using the `colors` parameter.  

plt.subplot(3, 3, 3)
colors = ['red', 'yellow', 'green', 'blue']
plt.pie(values, labels=categories, colors=colors, autopct='%1.1f%%')
plt.title('Pie Chart with Custom Colors')
```

```python
#### **4. Explode a Slice for Emphasis**  
#Pull out the **largest categorys** slice using the `explode` parameter to highlight it.

plt.subplot(3, 3, 4)
explode = (0, 0, 0.1, 0)  # Explode the 3rd slice (Cherry)
plt.pie(values, labels=categories, colors=colors, explode=explode, autopct='%1.1f%%')
plt.title('Pie Chart with Exploded Slice')
```

```python
#### **5. Add a Title to a Pie Chart**  
#Modify a **pie chart** by adding a **title**: `"Fruit Sales Distribution"`.  

plt.subplot(3, 3, 5)
plt.pie(values, labels=categories, colors=colors, autopct='%1.1f%%')
plt.title('Fruit Sales Distribution')
```

```python
#### **6. Rotate the Pie Chart (Start Angle Change)**  
#Change the **start angle** of the pie chart using `startangle=90` to rotate the chart.  

plt.subplot(3, 3, 6)
plt.pie(values, labels=categories, colors=colors, autopct='%1.0f%%', startangle=90)
plt.title('Rotated Pie Chart')
```

```python
#### **7. Convert a Pie Chart into a Donut Chart**  
#Modify a **pie chart** to look like a **donut chart** by adding a white circle at the center.

plt.subplot(3, 3, 7)
plt.pie(values, labels=categories, colors=colors, autopct='%1.0f%%', wedgeprops=dict(width=0.3))
plt.title('Donut Chart')
```

```python
#### **8. Create a Pie Chart with a Legend**  
#Display a **legend** alongside the **pie chart** using `plt.legend()`.

plt.subplot(3, 3, 8)
plt.pie(values, labels=categories, colors=colors, autopct='%1.0f%%')
plt.legend(categories, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))  # Place legend outside
plt.title('Pie Chart with Legend')
```

```python
#### **9. Use a Different Data Set for Pie Chart**  
#Create a **pie chart** using a different dataset:  
#- **Departments**: `['HR', 'IT', 'Marketing', 'Finance']`  
#- **Employees**: `[10, 25, 15, 20]`  

plt.subplot(3, 3, 9)
departments = ['HR', 'IT', 'Marketing', 'Finance']
employees = [10, 25, 15, 20]
plt.pie(employees, labels=departments, autopct='%1.1f%%')
plt.title('Employee Distribution by Department')

plt.tight_layout()  # Adjust subplot parameters for a tight layout
plt.show()  # Display the plots
```

```python
#### **10. Save a Pie Chart as an Image File**  
#Generate a **pie chart** and save it as `"pie_chart.png"` using `plt.savefig()`. 
 
plt.figure()
plt.pie(values, labels=categories, colors=colors, autopct='%1.1f%%')
plt.title('Fruit Sales Distribution')
plt.savefig('pie_chart.png')
```
