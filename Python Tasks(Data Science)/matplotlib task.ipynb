```python
 import matplotlib.pyplot as plt
import numpy as np
```

```python
#### **1. Create a Simple Line Plot**  
#Write a Python script to create a **line plot** with the following data:  
#- X values: `[1, 2, 3, 4, 5]`  
#- Y values: `[10, 20, 30, 40, 50]`  
#- Add **title and axis labels**.

plt.figure(figsize=(12, 8))  # Adjust figure size for better visualization
plt.subplot(3, 3, 1)  # Create subplots in a 3x3 grid
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
plt.plot(x, y)
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
#### **2. Plot a Bar Chart**  
#Create a **bar chart** showing sales data:  
#- Categories: `['Electronics', 'Clothing', 'Groceries']`  
#- Sales: `[15000, 25000, 18000]`  
#- Use different **colors** for each bar.

#plt.subplot(3, 3, 2)
categories = ['Electronics', 'Clothing', 'Groceries']
sales = [15000, 25000, 18000]
colors = ['blue', 'green', 'red']
plt.bar(categories, sales, color=colors)
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Sales')
```

```python
#### **3. Create a Scatter Plot**  
#Generate a **scatter plot** using random data for **X and Y coordinates**.  
#- Use `numpy` to generate **50 random points**.  
#- Color the points **red** and add a **title**.
 
plt.subplot(3, 3, 3)
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, color='red')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
#### **4. Generate a Histogram**  
#Create a **histogram** using **random normally distributed data** (1000 points).  
#- Use `numpy` to generate data.  
#- Set **bins=30**.  
#- Change the color to **purple**.

data = np.random.randn(1000)
plt.hist(data, bins=30, color='purple')
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
```

```python
#### **5. Customize a Plot (Title, Labels, and Grid)**  
#Modify a **line plot** by:  
#- Adding a **title**: `"Monthly Sales Trend"`  
#- Labeling the **X-axis** as `"Month"`  
#- Labeling the **Y-axis** as `"Sales"`  
#- Adding a **grid** to the plot.

plt.subplot(3, 3, 5)
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
plt.plot(x, y)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
```

```python
#### **6. Plot Multiple Lines on the Same Graph**  
#Create a **multi-line plot** with two sets of data:  
#- Line 1: `[1, 2, 3, 4] → [10, 15, 25, 35]` (blue, dashed)  
#- Line 2: `[1, 2, 3, 4] → [5, 10, 15, 20]` (red, dotted)  
#- Add **legend** for both lines.

plt.subplot(3, 3, 6)
x1 = [1, 2, 3, 4]
y1 = [10, 15, 25, 35]
x2 = [1, 2, 3, 4]
y2 = [5, 10, 15, 20]
plt.plot(x1, y1, linestyle='--', color='blue', label='Line 1')
plt.plot(x2, y2, linestyle=':', color='red', label='Line 2')
plt.legend()
plt.title('Multiple Lines')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
#### **7. Change Line Style and Markers**  
#Plot a **line graph** where:  
#- X values: `[1, 2, 3, 4, 5]`  
#- Y values: `[10, 30, 20, 40, 50]`  
#- Use **dashed lines** and **circle markers (`'o'`)**.  

plt.subplot(3, 3, 7)
x = [1, 2, 3, 4, 5]
y = [10, 30, 20, 40, 50]
plt.plot(x, y, linestyle='--', marker='o')
plt.title('Line Style and Markers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
### **8. Subplots: Display Multiple Graphs**  
#Create a figure with **two subplots**:  
#- **First subplot**: A bar chart  
#- **Second subplot**: A line plot  
#- Arrange subplots in a **1-row, 2-column format**.  

plt.subplot(3, 3, 8)
# Bar chart in first subplot
categories = ['A', 'B', 'C']
values = [10, 20, 15]
plt.bar(categories, values)
plt.title('Bar Chart')
plt.subplot(3, 3, 9)
# Line plot in second subplot
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
plt.plot(x, y)
plt.title('Line Plot')
```

```python
### **9. Create a Pie Chart**  
#Create a **pie chart** showing the percentage distribution of three products:  
#- Labels: `['Product A', 'Product B', 'Product C']`  
#- Values: `[40, 35, 25]`  
#- Add **colors** and **percentage labels**.  

plt.figure()  # Create a new figure for the pie chart
labels = ['Product A', 'Product B', 'Product C']
values = [30, 40, 30]
colors = ['red', 'green', 'blue']
plt.pie(values, labels=labels, colors=colors, autopct='%1.0f%%') 
plt.title('Pie Chart')

plt.tight_layout()  # Adjust subplot parameters for a tight layout
plt.show()  #display plot
```

```python
### **10. Save a Matplotlib Figure as an Image**  
#Create a **bar chart** and save it as `"sales_chart.png"` using `plt.savefig()`.

plt.figure()  # Create a new figure for the pie chart
labels = ['Product A', 'Product B', 'Product C']
values = [30, 40, 30]
colors = ['red', 'green', 'blue']
plt.pie(values, labels=labels, colors=colors, autopct='%1.0f%%') 
plt.title('Pie Chart')

plt.tight_layout()  # Adjust subplot parameters for a tight layout

plt.savefig('sales_chart.png') 
```
