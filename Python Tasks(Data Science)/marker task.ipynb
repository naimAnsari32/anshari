```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#### **1. Create a Basic Line Plot with Markers**  
#Plot a simple **line graph** with `x = [1, 2, 3, 4, 5]` and `y = [10, 15, 7, 20, 13]`, using **circle (`'o'`) markers** at each data point.  

plt.figure(figsize=(12, 6))  # Adjust figure size for better visualization
plt.subplot(3, 3, 1)  # Create subplots in a 3x3 grid
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 20, 13]
plt.plot(x, y, marker='o')  # Use circle markers
plt.title('Basic Line Plot with Markers')
```

```python
#### **2. Use Different Marker Styles in a Line Plot**  
#Plot **two different lines** in the same graph:  
#- First line: Use **triangle-up (`'^'`) markers**.  
#- Second line: Use **square (`'s'`) markers**. 
 
plt.subplot(3, 3, 2)
x1 = [1, 2, 3, 4, 5]
y1 = [10, 15, 7, 20, 13]
x2 = [1, 2, 3, 4, 5]
y2 = [12, 18, 9, 15, 10]
plt.plot(x1, y1, marker='^')  # Triangle-up markers
plt.plot(x2, y2, marker='s')  # Square markers
plt.title('Different Marker Styles')
```

```python
#### **3. Create a Scatter Plot with Star Markers**  
#Generate **50 random points** and display them in a **scatter plot** using **star (`'*'`) markers**.  

plt.subplot(3, 3, 3)
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, marker='*')  # Star markers
plt.title('Scatter Plot with Star Markers')
```

```python
#### **4. Change Marker Size in a Line Plot**  
#Modify a **line graph** to have **diamond (`'D'`) markers** with a **marker size of 12**. 

plt.subplot(3, 3, 4)
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 20, 13]
plt.plot(x, y, marker='D', markersize=6)  # Diamond markers, size 12
plt.title('Marker Size Change')

```

```python
#### **5. Customize Marker Colors**  
#Create a **line plot** where:  
#- The **inside color** of the marker is **yellow** (`markerfacecolor='yellow'`).  
#- The **border color** of the marker is **black** (`markeredgecolor='black'`).  

plt.subplot(3, 3, 5)
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 20, 13]
plt.plot(x, y, marker='o', markerfacecolor='yellow', markeredgecolor='black')
plt.title('Custom Marker Colors')
```

```python
#### **6. Use Different Markers for Multiple Lines**  
#Plot **three lines**, each with a different **marker style**:  
#- `'o'` (circle)  
#- `'p'` (pentagon)  
#- `'x'` (cross)  

plt.subplot(3, 3, 6)
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 7, 20, 13]
y2 = [12, 18, 9, 15, 10]
y3 = [8, 12, 15, 18, 12]
plt.plot(x, y1, marker='o')
plt.plot(x, y2, marker='p')
plt.plot(x, y3, marker='x')
plt.title('Multiple Lines, Different Markers')
```

```python
#### **7. Create a Multi-Line Graph with Legends**  
#Create a **multi-line graph** where each line has a **unique marker** and a **legend** for easy identification. 

plt.subplot(3, 3, 2)
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 7, 20, 13]
y2 = [12, 18, 9, 15, 10]
y3 = [8, 12, 15, 18, 12]
plt.plot(x, y1, marker='o', label='Line 1')
plt.plot(x, y2, marker='p', label='Line 2')
plt.plot(x, y3, marker='x', label='Line 3')
plt.legend()  # Show legend
plt.title('Multi-Line Graph with Legends')
```

```python
#### **8. Plot a Custom Styled Scatter Plot**  
#Modify a **scatter plot** to have:  
#- **Hexagon (`'H'`) markers**  
#- **Marker size of 100**  
#- **Color gradient based on values**  

plt.subplot(3, 3, 2)
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)  # Random colors for gradient
plt.scatter(x, y, marker='H', s=100, c=colors, cmap='viridis')  # Hexagon markers, size 100, color gradient
plt.title('Custom Styled Scatter Plot')
```

```python
#### **9. Add Grid and Titles to a Marked Plot**  
#Create a **line plot** with:  
#- **Markers for data points**  
#- **Title**: `"Sales Data Visualization"`  
#- **X-axis label**: `"Month"`  
#- **Y-axis label**: `"Revenue ($)"`  
#- **Grid lines enabled**  

plt.subplot(3, 3, 5)
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 20, 13]
plt.plot(x, y, marker='o')
plt.title('Sales Data Visualization')
plt.xlabel('Month')
plt.ylabel('Revenue ($)')
plt.grid(True)  # Enable grid lines

plt.tight_layout()  # Adjust subplot parameters for a tight layout
plt.show()  # Display the plots
```

```python
#### **10. Save a Matplotlib Figure with Markers**  
#Generate a **scatter plot** with **red cross (`'x'`) markers**, and save the figure as `"scatter_plot.png"`.

plt.figure()
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, marker='x', color='red')
plt.savefig('scatter_plot.png')  # Save the figure
```
